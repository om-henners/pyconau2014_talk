First import the libraries that I'm going to need

```python
import rasterio
import fiona
import shapely.geometry
import shapely.ops
import pyproj
import matplotlib.pyplot as plt
import cartopy
import numpy as np
import scipy.cluster.hierarchy
```

Then define some constants

```python
brisbane_convention_centre = (153.018374, -27.476484)

imagery_path = "./data/3V060921P0001382301A520000100142M_001603505.jpg"
tweets_path = "./data/big_tweets.shp"

wgs84 = pyproj.Proj(init="epsg:4326")
z56_s = pyproj.Proj(init="epsg:32756")

brisbane_convention_centre = pyproj.transform(wgs84, z56_s, *brisbane_convention_centre)

window = 1500 #metres radius

bbox = (
    brisbane_convention_centre[0] - window,
    brisbane_convention_centre[1] - window,
    brisbane_convention_centre[0] + window,
    brisbane_convention_centre[1] + window,
)
```

Read in a list of geometries (because in this case we're not too worried about the content of the tweets)

```python
with fiona.open(tweets_path) as source:
    records = source.filter(bbox=bbox)
    
    all_points = np.array(
        [rec["geometry"]["coordinates"] + (rec["properties"]["pyconau"], ) for rec in source]
    )
    
 
```

I can now cluster my data happily using scipy.cluster.hierarchy

```python
clusters = scipy.cluster.hierarchy.fclusterdata(all_points[:, :2], 100, criterion="distance", method="single")
```

And build some nice buffers around the clusters

```python
multipoints = []
print np.bincount(clusters)

for i, count in enumerate(np.bincount(clusters)):
    if count < 3:
        continue
    coords = all_points[clusters == i, :2]
    multipoints.append(shapely.geometry.asMultiPoint(coords).convex_hull.buffer(10))
```

2 warnings:

* The upper left corner of the image is the minimum x value and the *maximum* y value. Likewise the lower right is the maximum x value and the minimum y value
* Image extents in matplotlib are given as `(xmin, xmax, ymin, ymax)` tuples, not the `(xmin, ymin, xmax, ymax)` of the usual bounding box

```python
with rasterio.open(imagery_path) as source:
    upper_left = np.floor(~source.affine * (bbox[0], bbox[3])) #upper left is xmin, ymax
    lower_right = np.ceil(~source.affine * (bbox[2], bbox[1])) #lower right is xmax, ymin
    image = source.read_band(1, window=((upper_left[1], lower_right[1]), (upper_left[0], lower_right[0])))
    ul_bounds = source.affine * upper_left
    lr_bounds = source.affine * lower_right
    img_bounds = (ul_bounds[0], lr_bounds[0], lr_bounds[1], ul_bounds[1])
    print img_bounds
```

And finally we can make a map - there's actually a whole lot that we can add to the map using cartopy as well, so of which are commented below (as none of them would show up!)

```python

proj = cartopy.crs.epsg(32756)
plt.figure(figsize=(24, 24))
ax = plt.axes(projection=proj)

# ax.add_feature(cartopy.feature.OCEAN)
ax.gridlines()
# ax.add_feature(cartopy.feature.COASTLINE)
# ax.add_feature(cartopy.feature.NaturalEarthFeature(category='physical',
#                                                   name='ocean',
#                                                   scale='50m'), facecolor='lightblue')
# ax.add_feature(cartopy.feature.NaturalEarthFeature(category='physical',
#                                                   name='coastline',
#                                                   scale='50m'), facecolor='none')
# ax.add_feature(cartopy.feature.NaturalEarthFeature(
#         category='cultural',
#         name='admin_1_states_provinces_lines',
#         scale='50m'),
#         facecolor='none', edgecolor='gray')

ax.imshow(image, extent=img_bounds, transform=proj, origin='upper', cmap='gray')
ax.scatter(all_points[:,0], all_points[:, 1], transform=proj, zorder=2)
ax.add_geometries(multipoints, proj, facecolor="yellow", edgecolor="orange", alpha=0.3)

ax.set_extent(img_bounds, proj)
```

So why in this notebook did I worry about adding in the #pyconAU column? Becuase I can do some spatial analysis to determine if the presence is spatially dependent ([pysal](http://pysal.readthedocs.org/en/latest/index.html) is awesome btw)

```python
import pysal
```

```python
weights = pysal.pysal.threshold_binaryW_from_array(all_points[:, :2], 100) #Assume spatial dependence @ 100m
```

```python
moran = pysal.Moran(all_points[:, 2], weights)
print moran.I #not very correlated
```

```python
moran_local = pysal.Moran_Local(all_points[:, 2], weights)
```

Filter out points that are isolated, and work out which points are significant (pseudo p-score)

```python
island_filter = np.ones(len(all_points))
island_filter[weights.islands] = 0
significant_points = all_points[:, :2][np.where((island_filter > 0) & (moran_local.p_sim < 0.05))]
len(significant_points) #So there's a lot of points which don't fit the normailty assumption
```

```python
plt.figure(figsize=(12, 12))
ax = plt.axes(projection=proj)

# ax.add_feature(cartopy.feature.OCEAN)
ax.gridlines()
ax.imshow(image, extent=img_bounds, transform=proj, origin='upper', cmap='gray')
ax.scatter(all_points[:,0], all_points[:, 1], transform=proj, zorder=2)
ax.scatter(significant_points[:,0], significant_points[:, 1], transform=proj, zorder=3, s=100, c='yellow')
ax.set_extent(img_bounds, proj)
```
