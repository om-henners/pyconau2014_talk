Assume you want to know how a projection is going to distort your data. A good way to look at this is with Tissot's index (http://en.wikipedia.org/wiki/Tissot%27s_indicatrix). It's easy to construct in Python as well.

```python
import numpy as np
import matplotlib.pyplot as plt
from shapely import geometry, ops
import cartopy
```

```python
spacing = 45.
x, y = np.arange(-180, 180, spacing) + spacing / 2, np.arange(-90, 90, spacing) + spacing / 2
x_view, y_view = np.meshgrid(x, y, indexing='ij')
buffers = [
    geometry.Point(x_view[i, j], y_view[i, j]).buffer(spacing / 3)
    for i in xrange(len(x))
    for j in xrange(len(y))
]
```

```python
geodetic = cartopy.crs.PlateCarree() # approximate equivalent to geodetic = cartopy.crs.Geodetic()
tissot = cartopy.feature.ShapelyFeature(buffers, geodetic)
```

```python
for proj in [
        cartopy.crs.PlateCarree(), cartopy.crs.Mercator(), cartopy.crs.Mollweide(),
        cartopy.crs.InterruptedGoodeHomolosine(), cartopy.crs.SouthPolarStereo(),
        cartopy.crs.TransverseMercator(), #etc...
]:
    plt.figure(figsize=(12, 12))
    ax = plt.axes(projection=proj)
    ax.add_feature(cartopy.feature.COASTLINE)
    ax.add_feature(tissot, edgecolor='darkblue', facecolor='blue', alpha=0.3)
    plt.show()
```
